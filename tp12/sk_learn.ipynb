```python
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
```

```python
np.random.seed(0)
```

```python
m = 100
X = np.linspace(0, 10, m).reshape(m, 1)
y = X + np.random.randn(m, 1)
plt.scatter(X, y)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
reg.score(X, y)
```

```python
predictions = reg.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, predictions, c='r')
```
